```python
import pandas as pd

# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = Gender_df['long_hair'].quantile(0.25)
Q3 = Gender_df['long_hair'].quantile(0.75)

# Compute IQR
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
df_cleaned = Gender_df[(Gender_df['long_hair'] >= lower_bound) & (Gender_df['long_hair'] <= upper_bound)]

# Display results
print(f"Original dataset size: {Gender_df.shape[0]}")
print(f"After removing outliers: {df_cleaned.shape[0]}")
```

```python
Gender_df=df_cleaned
```

```python
for i in Gender_df.select_dtypes(include=np.number).columns:
  sns.boxplot(Gender_df[i])
  plt.show()
```

```python


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression # Model
from sklearn.tree import DecisionTreeClassifier # Model
from sklearn.ensemble import RandomForestClassifier # Model
```

```python
from sklearn.metrics import classification_report, accuracy_score
```

```python
x=Gender_df.drop(['gender'],axis=1)
y=Gender_df['gender']
print(x.head())
print(y.head())
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=55)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
decision_tree_model = DecisionTreeClassifier()
decision_tree_model.fit(x_train, y_train)
```

```python
decision_tree_model.score(x_test,y_test)
```

```python
ds_predicted=decision_tree_model.predict(x_test)
accuracy_ds = accuracy_score(y_test,ds_predicted)
accuracy_ds
```

```python

from sklearn.ensemble import AdaBoostClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier


from sklearn.model_selection import train_test_split
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
random_forest_model = RandomForestClassifier()
random_forest_model.fit(x_train, y_train)
```

```python
random_forest_model.score(x_test,y_test)
```

```python
rf_predicted=random_forest_model.predict(x_test)
accuracy_rf = accuracy_score(y_test,rf_predicted)
accuracy_rf
```

# AdaBoost 

```python
ada = AdaBoostClassifier()
ada.fit(x_train, y_train)
y_pred = ada.predict(x_test)

print('AdaBoostClassification report ',classification_report(y_test,y_pred))
print('adaboost Accuracy ',accuracy_score(y_test,y_pred))
```

```python
ada.score(x_test,y_test)
```

```python
ada = AdaBoostClassifier(n_estimators=100,random_state=100)
ada.fit(x_train, y_train)
y_pred = ada.predict(x_test)

print('AdaBoost Classification report ',classification_report(y_test,y_pred))
print('Adaboost Accuracy ',accuracy_score(y_test,y_pred))
```

# Gradient Boosting

```python
gbc = GradientBoostingClassifier(n_estimators=100,random_state=100)
gbc.fit(x_train, y_train)
y_pred = gbc.predict(x_test)

print('Gradient Boosting Classifier ',classification_report(y_test,y_pred))
print('gbc Accuracy ',accuracy_score(y_test,y_pred))
```

# LGBM Classifier

```python
lgbm = LGBMClassifier(n_estimators=100,random_state=100)
lgbm.fit(x_train, y_train)
y_pred = lgbm.predict(x_test)

print('LGBM Classifier ',classification_report(y_test,y_pred))
print('lgbm Accuracy ',accuracy_score(y_test,y_pred))
```

# KNeighborsClassifier

```python

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)

print('LGBM Classifier ',classification_report(y_test,y_pred))
print('lgbm Accuracy ',accuracy_score(y_test,y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(x_train, y_train)
y_pred = gnb.predict(x_test)

print('GaussianNB',classification_report(y_test,y_pred))
print('gnb Accuracy ',accuracy_score(y_test,y_pred))
```
